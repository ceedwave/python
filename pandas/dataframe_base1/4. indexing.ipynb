#### 4-1. 위치기반 인덱싱

```python
#iloc = 위치기반 메소드
import pandas as pd
titanic = pd.read_csv("Video_Lecture_NBs/titanic.csv")
```

```python
titanic
```

```python
# 행, 열, 그리고 요소들의 위치를 선택하는 방법 1
# zero based indexing 
# 첫 번째 열의 경우 index로 column 이 아님

titanic[["survived"]]
```

```python
# 행, 열, 그리고 요소들의 위치를 선택하는 방법 2
# negative indexing 
# 마지막 열부터 -1, -2 순차적으로 음수로 위치값을 나타낼 수 있다, 행의 경우도 동일

titanic.iloc[[-1]]
```

```python
summer = pd.read_csv("Video_Lecture_NBs/summer.csv", index_col="Athlete")
```

```python
summer
```

```python
summer.info()
```

```python
# summer.iloc[0] > pandas 시리즈
type(summer.iloc[0])
```

```python
# summer.iloc[[0]] > 자료형 (데이터 프레임)
type(summer.iloc[[0]])
```

```python
# 슬라이싱
summer.iloc[[-1,-15,-111]] 

# 범위 슬라이싱의 경우 시작 숫자는 포함 마지막 숫자는 미포함
# 1 ~ 11까지의 숫자를 슬라이싱
#summer.iloc[1:12]

# 음수의 경우 
# 마지막 5개의 행을 출력 = summer.iloc[-5:]
# 마지막 5번째부터 마지막행까지는 제외 summer.iloc[-5:-1], 4개의 행만 출력
#summer.iloc[-5:-1]

# 모든행의 출력
#summer.iloc[:]
```

```python
summer.head(10)
```

```python
# 좌측은 행, 우측은 열 위치값을 넣으면 된다.
summer.iloc[0,4]
```

```python
# 슬라이싱을 이용한 행렬 선택
summer.iloc[:, 4]
# summer.iloc[:, 4].equals(summer.Country) 이퀄 함수로 슬라이싱 표기법과 닷(점) 표기법이 일치하는지 확인할 수 있다.
# summer["Country"] 칼럼명을 이용한 선택
```

#### 4-2. 레이블 기반 인덱싱

```python
# loc = 레이블 기반 인뎅싱
summer
```

```python
# 해당 행의 경우 1개의 행이기 때문에 pandas 시리즈 1차원 배열로 노출됨
summer.loc["DRIVAS, Dimitrios"]
```

```python
# 해당 행의 경우 복수의 행이기 때문에 데이터 프레임으로 노출됨
summer.loc["PHELPS, Michael"]
```

```python
# 해당 열의 경우에도 1개의 컬럼이기 때문에 pandas 시리즈 1차원 배열로 노출됨
summer.loc["PHELPS, Michael", "Medal"]
```

```python
# 해당 열의 경우 복수의 열이기 때문에 데이터 프레임으로 노출됨
# 복수의 컬럼을 슬라이싱할 경우 리스트 형태로 작성해줘야함
summer.loc["PHELPS, Michael", ["Medal", "Event"]]
```

```python
# 복수의 행열을 입력할 경우 리스트 형태로 작성해줘야함
summer.loc[["PHELPS, Michael", "LEWIS, Carl"], ["Medal", "Event"]]
```

```python
# 전체 행에 대해서 불러오기
summer.loc[:, ["Medal", "Event"]]
```
