```python
def solution(n):
    result = 0
    for i in range(1, n, 2):
        result += i
    return result

n = 15
answer = solution(n)
print(answer)
```

```python
def solution(numbers):
    answer = 0
    a = len(numbers)
    for i in range(numbers):
        answer += i
        answer / a
        
    return answer
    
a = solution(10)
print(a)
```

```python
# 리스트는 대괄호
# 리스트 추가

list = [1,2,3,4,5,6]
list.append(25) # append 함수 (값)
print(list)

# 리스트 삭제

li = [1,2,3,4,5,76]
del(li[5]) # del 함수 [위치]
print(li)

li2 = [5,6,3,4,5,76]
li2.remove(3) # Remove 함수 (값)
print(li2)
```

```python
# 딕셔너리는 중괄호
dict = {
    'one' : 1,
    'two' : 2,
    'three' : 3,
    'four' : 4
}

# 딕셔너리 추가
dict['five'] = 5 # [이름] : 값
dict

# 딕셔너리 수정
dict['two'] = 78 # 추가하는 것과 동일한 형식으로 이미 존재하는 값을 덮어쓰는 방식
dict

# 딕셔너리 삭제
del (dict['one']) # 리스트의 del 사용법과 동일 대신 이름을 삽입
dict

pop = dict.pop('four') # pop 함수로 삭제할 때 특징은 삭제된 값을 반환해준다.
print(pop)

dict
```

```python
def check_and_clear(box):
    if "불량품" in box.keys():
        box = box.clear()
    print("불량품이 있으면 box를 clear합니다.")

box1 = {"불량품" : 10}
check_and_clear(box1)
# {}가 출력되어야합니다.
print(box1)

box2 = {"정상품": 10}
check_and_clear(box2)
# {"정상품": 10}가 출력되어야합니다.
print(box2)
```
